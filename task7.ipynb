```python
import sqlite3
conn = sqlite3.connect("sales_table.db")

cur = conn.cursor()
```

```python
query = """
SELECT
    product,
    SUM(quantity)    AS total_qty,
    SUM(quantity * price) AS revenue
FROM sales
GROUP BY product
"""
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_sql_query(query, conn)

df
```

```python
ax = df.plot(
    kind='bar',
    x='product',
    y='revenue',
    legend=False
)
ax.set_xlabel('Product')
ax.set_ylabel('Revenue')
ax.set_title('Revenue by Product')
plt.tight_layout()
plt.show()
```
